```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create dataset
X, y = make_regression(n_samples=100, n_features=2, noise=0.1)
# Train, Test, Predict using LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
reg = LinearRegression().fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

```python
# 3D Scatter Plot without grid
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], y, c='r', marker='o')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')
ax1.set_zlabel('Target')
ax1.set_title("Without Grid by Aditya Mishra")
ax1.grid(False) # Disable grid
```

```python
fig = plt.figure(figsize=(12, 6))
ax2 = fig.add_subplot(122, projection='3d')
# Scatter actual data points
ax2.scatter(X[:, 0], X[:, 1], y, c='r', marker='o', label='Data points')
# Labels and title
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')
ax2.set_zlabel('Target')
ax2.set_title("With Grid and Regression Hyperplane")
ax2.grid(True)
# Create meshgrid for hyperplane
x_surf, y_surf = np.meshgrid(
    np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
    np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
)
# Calculate z values based on your regression coefficients
z_surf = reg.intercept_ + reg.coef_[0] * x_surf + reg.coef_[1] * y_surf
# Plot the regression hyperplane with transparency
ax2.plot_surface(x_surf, y_surf, z_surf, color='blue', alpha=0.3, rstride=10, cstride=10)
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax2 = fig.add_subplot(111, projection='3d')
# Create meshgrid for two features (X[:,0] and X[:,1])
x_surf, y_surf = np.meshgrid(
    np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
    np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
)
# Calculate predicted z values on the grid using the linear model equation
z_surf = reg.intercept_ + reg.coef_[0] * x_surf + reg.coef_[1] * y_surf
# Plot the surface
ax2.plot_surface(x_surf, y_surf, z_surf, color='blue', alpha=0.3, rstride=10, cstride=10)
# Optionally plot the actual data points in 3D
ax2.scatter(X[:, 0], X[:, 1], y, color='red')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')
ax2.set_zlabel('Target')
plt.show()
```

```python
# Calculate errors and r2_score
print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred):.2f}\
\nMean square Error: {mean_squared_error(y_test, y_pred):.2f}\
\nR2 Score: {r2_score(y_test, y_pred):.2f}")
print(f"Intercept: {reg.intercept_:.2f}, Coefficients: {reg.coef_}")
```
